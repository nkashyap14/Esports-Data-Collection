```python
aggregate_kill_data = []
aggregate_nade_data = []
aggregate_flash_data = []
aggregate_round_data = []
aggregate_damage_data = []
```

```python
for x in demo_parsers:
    aggregate_kill_data.append(x._parse_kills(return_type="df"))
    aggregate_nade_data.append(x._parse_grenades(return_type="df"))
    aggregate_flash_data.append(x._parse_flashes(return_type="df"))
    aggregate_round_data.append(x._parse_rounds(return_type="df"))
    aggregate_damage_data.append(x._parse_damages(return_type="df"))
```

```python
def concat_data(aggregate_data):
    final_df = None
    for x in aggregate_data:
        final_df = pandas.concat([final_df, x], ignore_index=True)
    return final_df
```

```python
final_kills = concat_data(aggregate_kill_data)
final_nades = concat_data(aggregate_nade_data)
final_flash = concat_data(aggregate_flash_data)
final_round = concat_data(aggregate_round_data)
final_damage = concat_data(aggregate_damage_data)
final_data = concat_data(aggregate_damage_data)
```

```python
def calculateside_volume_flashbangs(players, nades, side):
    counts = {}
    for x in players:
         counts[x] = len(final_nades[final_nades['PlayerSide'] == side][final_nades['GrenadeType'] == 'Flashbang'][final_nades['PlayerName'] == x].index)
    return counts
```

```python
def calculateRoundsPlayed(players, rounds):
    counts = {}
    for x in players:
        for y in rounds:
            if x not in counts.keys():
                counts[x] = 0
            if(x in y.AttackerName.value_counts()):
                counts[x] += y.RoundNum.iloc[-1]
    return counts
```

```python
def calculateTotalKills(players, kills):
    counts = {}
    for x in players:
        counts[x] = len(final_kills[final_kills['AttackerName'] == x].index)
    return counts
```

```python
def calculateVolumeHEDamage(players, nades):
    counts = {}
    for x in players:
        counts[x] = nades[nades['AttackerName'] == x][nades['Weapon'] == 'HE Grenade'].HpDamageTaken.sum()
    return counts
```

```python
def calculateVolumeHEDamageSide(players, nades, side):
    counts = {}
    for x in players:
        counts[x] = nades[nades['AttackerName'] == x][nades['Weapon'] == 'HE Grenade'][nades['AttackerSide'] == side].HpDamageTaken.sum()
    return counts
```

```python
def calculateKDDifferentialByPlaces(players, kills):
    places = []
    temp = kills.copy()
    for x in temp.MapName.unique():
        temp2 = temp[temp['MapName'] == x]
        for y in temp2.VictimAreaName.unique():
            places.append(temp2.MapName.unique()[0] + ":" + y)
    ret = {}
    for y in players:
        temp = {}
        for x in places:
            maps = x.split(':')[0]
            place = x.split(':')[1]
            k = len(kills[kills['MapName'] == maps][kills['AttackerName'] == y][kills['AttackerAreaName'] == place].index)
            d = len(kills[kills['MapName'] == maps][kills['VictimName'] == y][kills['VictimAreaName'] == place].index)
            #kills = 
            temp[x] = k - d
        ret[y] = temp
    return ret
```

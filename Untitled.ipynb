```python
grim_liquid_lineup = "/mnt/d/Grim/Liquid"
grim_triumph_lineup = "/mnt/d/Grim/Triumph"
```

```python
from csgo.parser import DemoParser
from csgo.analytics import stats
import numpy
import pandas
import os
```

```python
def getDemoPaths(directory):
    demoPaths = []
    for filename in os.listdir(directory):
        demoPaths.append(os.path.join(directory, filename))
    return demoPaths
```

```python
def generateDemoParsers(paths):
    demo_parsers = []
    for x in paths:
        demoParser = DemoParser(demofile=x, demo_id = x.split('/')[-1].split('.')[0].split('_')[0], parse_rate=128)
        demo_parsers.append(demoParser)
    print("# Demo Parsers: " + str(len(demo_parsers)))
    return demo_parsers
```

```python
def generateData(parsers):
    killData, roundData, damageData, frameData, playerFrames, bombEvents, grenadeData, flashData = [], [], [], [], [], [], [], []
    for x in parsers:
        x.parse()
        x.clean_rounds()
        killData.append(x._parse_kills())
        roundData.append(x._parse_rounds())
        damageData.append(x._parse_damages())
        frameData.append(x._parse_frames())
        playerFrames.append(x._parse_player_frames())
        bombEvents.append(x._parse_bomb_events())
        grenadeData.append(x._parse_grenades())
        flashData.append(x._parse_flashes())
    return ( killData, roundData, damageData, frameData, playerFrames, bombEvents, grenadeData, flashData)
```

```python
liquidKillData, liquidRoundData, liquidDamageData, liquidFrameData, liquidPlayerFrames, liquidBombEvents, liquidGrenadeData, liquidFlashData = generateData(generateDemoParsers(getDemoPaths(grim_liquid_lineup)))
```

```python
triumphKillData, triumphRoundData, triumphDamageData, triumphFrameData, triumphPlayerFrames, triumphBombEvents, triumphGrenadeData, triumphFlashData = generateData(generateDemoParsers(getDemoPaths(grim_triumph_lineup)))
```
